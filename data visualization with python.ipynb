```python
x=(1,3,7)
y=(1,6,3)
z=(7,2,4)
plt.plot(x,y)
plt.plot(x,y,'o')
plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("y")
plt.title("test")
plt.legend(["x-y line", "x-y line","x-z"])
plt.show
```

```python
x1=(1,3,5)
y1=(9,6,3)
x2=(9,6,1)
y2=(5,3,6)
plt.plot(x1,y1, linestyle = 'dashed', color = 'orange', linewidth = 3, marker = 'o',alpha = .9, markerfacecolor='b')
plt.plot(x2,y2, linestyle='-.')
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["line1", "line2"])
plt.title("test2")
plt.show()
```

```python
x1=(1,5,10,4)
y1=(3,6,9,1)
x2=(3,6,7)
y2=(1,5,10)
x3 = (1,4,2,7)
y3=(4,2,6,8)
plt.plot(x1,y1 , label = 'x1-y1 line', linestyle = 'dashed', linewidth = 3,alpha = .5, marker ='o', markerfacecolor = 'orange', color = 'red' )
plt.plot(x2,y2, label = "x2-y2 line", color='blue', linestyle = 'dashed', linewidth = 1, marker='o', markerfacecolor = 'black')
plt.plot(x3,y3, label = 'x3-y3 line', color='green', marker = 'x', markerfacecolor='black', linestyle='-.')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Lines")
plt.show()
```

```python
newdata2 = pd.read_csv('countries.csv')
```

```python
india = newdata2[newdata2.country == 'India']
```

```python
india
```

```python
us = newdata2[newdata2.country == 'United States']
```

```python
us
```

```python
plt.plot(india.year, india.population/10**6, label='India', color='green', linestyle='-.', marker='o', markerfacecolor='orange')
plt.plot(us.year, us.population/10**6, color='pink', marker='o', markerfacecolor='black')
plt.xlabel('year')
plt.ylabel('population(million)')
plt.legend()
plt.show()
```

```python
#for growth rate we need initial year population-
india.population.iloc[0]
```

```python
us.population.iloc[0]
```

```python
china=newdata2[newdata2.country == 'China']
```

```python
china
```

```python
# final growth rate line-plot
plt.plot(india.year, india.population/india.population.iloc[0]*100,label='India', marker='o')
plt.plot(us.year, us.population/us.population.iloc[0]*100, label='us', marker='o')
plt.plot(china.year, china.population/china.population.iloc[0]*100, label='china', marker='o')
plt.xlabel('year')
plt.ylabel('population growth rate(first year=100)%')
plt.legend()
plt.show()
```

```python
#scatter-plot
```

```python
# final growth rate line-plot
plt.scatter(india.year, india.population/india.population.iloc[0]*100,label='India', marker='o')
plt.scatter(us.year, us.population/us.population.iloc[0]*100, label='us', marker='o')
plt.scatter(china.year, china.population/china.population.iloc[0]*100, label='china', marker='o')
plt.xlabel('year')
plt.ylabel('population growth rate(first year=100)%')
plt.legend()
plt.show()
```

```python
#Pie-chart
```

```python
activities=['sleep', 'study', 'mobile', 'eat', 'waste' ]
slices = [9, 4, 3, 1, 7]
colors = ['r','y','b','g','y']
plt.pie(slices, colors=colors,labels=activities, startangle=90, radius=1.3, explode=(0,0,0,0,.1),autopct = '%1.1f%%')
plt.legend()
plt.show()
```
