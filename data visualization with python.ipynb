```python
# Customization of Plots-
# example - color='green', linestyle='dashed', linestyle='-.', linewidth = 3, marker='o', markerfacecolor='blue', markersize=12, aplha = .9
```

```python
x1=(1,3,5)
y1=(9,6,3)
x2=(9,6,1)
y2=(5,3,6)
plt.plot(x1,y1, linestyle = 'dashed', color = 'orange', linewidth = 3, marker = 'o',alpha = .9, markerfacecolor='b')
plt.plot(x2,y2, linestyle='-.')
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["line1", "line2"])
plt.title("test2")
plt.show()
```

```python
x1=(1,5,10,4)
y1=(3,6,9,1)
x2=(3,6,7)
y2=(1,5,10)
x3 = (1,4,2,7)
y3=(4,2,6,8)
plt.plot(x1,y1 , label = 'x1-y1 line', linestyle = 'dashed', linewidth = 3,alpha = .5, marker ='o', markerfacecolor = 'orange', color = 'red' )
plt.plot(x2,y2, label = "x2-y2 line", color='blue', linestyle = 'dashed', linewidth = 1, marker='o', markerfacecolor = 'black')
plt.plot(x3,y3, label = 'x3-y3 line', color='green', marker = 'x', markerfacecolor='black', linestyle='-.')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Lines")
plt.show()
```

```python
#for growth rate we need initial year population-
india.population.iloc[0]
```

```python
us.population.iloc[0]
```

```python
china=newdata2[newdata2.country == 'China']
```

```python
china
```

```python
# final growth rate line-plot
plt.plot(india.year, india.population/india.population.iloc[0]*100,label='India', marker='o')
plt.plot(us.year, us.population/us.population.iloc[0]*100, label='us', marker='o')
plt.plot(china.year, china.population/china.population.iloc[0]*100, label='china', marker='o')
plt.xlabel('year')
plt.ylabel('population growth rate(first year=100)%')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt 

# heights of bars
# y-
height = [10, 24, 36, 40, 5] 


# x-coordinates of left sides of bars 
# x-
left = [1, 2, 3, 4, 5] 
# left = range(len(height))


# left = [4,6,8,10,12]
# labels for bars 
tick_label = ['one', 'two', 'three', 'four', 'five'] 

# plotting a bar chart 
plt.bar(left, height,  width = 0.8, color = ['red', 'green']) 
# plt.bar(left, height, tick_label = tick_label, width = 0.8, color = ['red', 'green']) 

# naming the x-axis 
plt.xlabel('x - axis') 
# naming the y-axis 
plt.ylabel('y - axis') 
# plot title 
plt.title('Bar chart!') 

# function to show the plot 
plt.show()

```
